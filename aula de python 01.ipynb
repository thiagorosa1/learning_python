utilizando função for

```
for fonte in fontes:
  print(fonte)
  print("...")
print("ACABOU A LISTA")  
```

```
def imprime_fontes(fontes_que_quero_imprimir):
  print("lista de fontes disponível")
  for fonte in fontes_que_quero_imprimir:
    print(fonte)
```

```
imprime_fontes(fontes)
```
