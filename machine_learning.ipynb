```python
#importar biblioteca
import pandas as pd
```

```python
#carregar arquivo e coloca-lo em uma variável
arquivo = pd.read_csv('wine_dataset.csv')
arquivo.tail()
```

```python
#verificar o tipo de dados da coluna
arquivo.dtypes
```

```python
#alterar tipo de dado para binário
arquivo['style'] = arquivo['style'].replace('red',0)
arquivo['style'] = arquivo['style'].replace('white',1)
```

```python
#virificar se os valores foram substituidos
arquivo.tail()
```

```python
#seoparando entre variaveis preditora  e alvo
y = arquivo['style']
x = arquivo.drop('style', axis = 1)
```

```python
#verificar o que esta na variavel 'y'
print(y)
```

```python
#verificar o que esta na variavel 'x
x
```

```python
#criar o conjunto de dados para treino e teste
from sklearn.model_selection import train_test_split

x_treino, x_teste, y_treino, y_teste = train_test_split(x, y, test_size = 0.3)

```

```python
#verificar o temanho da variavel de treino
x_treino.shape
```

```python
#criar modelo de machine learning
from sklearn.ensemble import ExtraTreesClassifier
modelo = ExtraTreesClassifier()
modelo.fit(x_treino, y_treino)
```

```python
#imprimir resultado comparando a teste os dados reais
resultado = modelo.score(x_teste, y_teste)
print('Acuracia:', resultado)
```
